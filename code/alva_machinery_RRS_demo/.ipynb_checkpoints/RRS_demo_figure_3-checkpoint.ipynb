## 1.a random_seeding for getting a rough profile of the neurite_world
## 1.b selected_seeding for the detial of the neurite_world

```python
section_name = '_random_seeding'
###
### image_array_loading +++
image_name = '_zoomX'
channel_color = channel_color_order[1]
file_suffix = '.npy'
mm = np.load(os.path.join(saving_dir_path, file_name + image_name + channel_color + file_suffix))
mmm = np.copy(mm)
total_pixel_y, total_pixel_x = mmm.shape
### image_array_loading ---
###
### AlvaHmm_parameter +++
total_node = int(6)
total_path = int(3+3) 
node_r = int(5)
node_angle_max = 90 * (np.pi / 180)
### AlvaHmm_parameter ---
###
### seeding +++
### random_zone +++
seed_xx = np.array([])
seed_yy = np.array([])
total_seed = 2048#int(mmm.size / (10 * node_r * total_node))
###
random_xx = np.random.random(total_seed)
random_yy = np.random.random(total_seed)
boundary_width = node_r
boundary_cut_y = 0#total_pixel_y - 280
for i in range(total_seed):
    x = random_xx[i]
    x_random_range = total_pixel_x - boundary_width
    x0 = int(x_random_range * x) + boundary_width
    ##
    seed_xx = np.append(seed_xx, x0)
    ### start_y
    y = random_yy[i]
    y_random_range = total_pixel_y - boundary_width - boundary_cut_y
    y0 = int(y_random_range * y) + boundary_width
    seed_yy = np.append(seed_yy, y0) 
### random_zone ---
### seeding ---
###############
### AlvaHMM +++
chain_HMM = alva_MCMC.chain_HMM_node(mmm, 
                                     total_node,
                                     total_path,
                                     node_r,
                                     node_angle_max,
                                     seed_xx,
                                     seed_yy,
                                     seed_angle = None, 
                                     chain_level = 1.16,
                                     )
###
real_chain_ii, real_chain_xx, real_chain_yy = chain_HMM[0:3]
seed_node_xx, seed_node_yy = chain_HMM[3:5]
demo_scope = chain_HMM[-1]
###
seed_node_path_xx, seed_node_path_yy, seed_node_path_pp = demo_scope[0:3]
seed_node_path_path0max = demo_scope[3]
seed_node_path_path0_xx, seed_node_path_path0_yy, seed_node_path_path0_pp = demo_scope[4:-2]
seed_node_path_path0_xx_seed, seed_node_path_path0_yy_seed = demo_scope[-2:]
### AlvaHMM ---
###############
### plotting
figure_name = '{:}{:}'.format(section_name, image_name)
file_suffix = '.png'
save_figure = os.path.join(saving_dir_path, file_name + figure_name + file_suffix)
AlvaFontSize = 21
wX = 2
wY = 1
wxx = 9 # window_width
figure = plt.figure(figsize = (wX*wxx, wY*wxx*(raw_pixel_y / raw_pixel_x)))
for wn in iWatcher(range(wX*wY)):
    if (wn == 0):
        window = figure.add_subplot(wY, wX, wn + 1)
        window.set_title('total random seeds = {:}'.format(total_seed), fontsize = AlvaFontSize)
        ### all seeding_point and their tracking
        for i in range(total_seed):
            window.plot(seed_node_xx[i], seed_node_yy[i],
                        '.', markersize = node_r / 1, 
                        markerfacecolor = 'none', markeredgecolor = 'black', markeredgewidth = 0.25)
            window.plot(seed_node_xx[i, 0], seed_node_yy[i, 0],
                        '*', markersize = 10, markerfacecolor = 'none', markeredgecolor = 'blue')
        ### real_chain
        for i in range(real_chain_ii.shape[0]):
            window.plot(real_chain_xx[i][real_chain_ii[i]], real_chain_yy[i][real_chain_ii[i]],
                        'o', markersize = node_r / 0.8, markerfacecolor = 'none',
                        markeredgecolor = 'red', markeredgewidth = 0.25)
        ###
        window.imshow(mmm, cmap = channel_color + 's', interpolation = 'nearest')
        window.set_xlabel('X (pixel)', fontsize = AlvaFontSize*0.8)
        window.set_ylabel('Y (pixel)', fontsize = AlvaFontSize*0.8)
        window.set_xticklabels(np.int64(window.get_xticks()), fontsize = AlvaFontSize*0.8, rotation = 0) 
        window.set_yticklabels(np.int64(window.get_yticks()), fontsize = AlvaFontSize*0.8)
        #window.grid(True)
        ###
    if (wn == 1):
        ### selected seeding +++
        selected_seed_xx = np.array([])
        selected_seed_yy = np.array([])
        for i in range(real_chain_ii.shape[0]):
            if len(real_chain_ii[i]) >= int(total_node / 2):
                selected_seed_xx = np.append(selected_seed_xx, real_chain_xx[i][real_chain_ii[i]][1])
                selected_seed_yy = np.append(selected_seed_yy, real_chain_yy[i][real_chain_ii[i]][1])
        ### selected seeding ---
        window = figure.add_subplot(wY, wX, wn + 1)
        window.set_title('selected seeds = {:}'.format(len(selected_seed_xx)), fontsize = AlvaFontSize)
        ### 
        window.plot(selected_seed_xx, selected_seed_yy,
                    '*', markersize = 10, markerfacecolor = 'none', markeredgecolor = 'blue')
        ###
        window.imshow(mmm, cmap = channel_color + 's', interpolation = 'nearest')
        window.set_xlabel('X (pixel)', fontsize = AlvaFontSize*0.8)
        window.set_ylabel('Y (pixel)', fontsize = AlvaFontSize*0.8)
        window.set_xticklabels(np.int64(window.get_xticks()), fontsize = AlvaFontSize*0.8, rotation = 0) 
        window.set_yticklabels(np.int64(window.get_yticks()), fontsize = AlvaFontSize*0.8)
        #window.grid(True)    
###
figure.tight_layout()
plt.savefig(save_figure, dpi = 128, bbox_inches = 'tight')
plt.show()
#####################
```

```python
section_name = '_selected_seeding'
###
### image_array_loading +++
image_name = '_zoomX'
channel_color = channel_color_order[1]
file_suffix = '.npy'
mm = np.load(os.path.join(saving_dir_path, file_name + image_name + channel_color + file_suffix))
mmm = np.copy(mm)
total_pixel_y, total_pixel_x = mmm.shape
### image_array_loading ---
###
### AlvaHmm_parameter +++
total_path = int(12) 
node_r = int(5)
total_node = 50#int(total_pixel_y / node_r / 5)
node_angle_max = 90 * (np.pi / 180)
### AlvaHmm_parameter --- 
###
### seeding +++
seed_xx = np.copy(selected_seed_xx)
seed_yy = np.copy(selected_seed_yy)
###
seed_aa = np.array([])
for i in range(len(seed_xx)):
    ### angle
    start_angle = np.random.random(1) * 360 * (np.pi / 180)
    seed_aa = np.append(seed_aa, start_angle)
### seeding ---
###############
### AlvaHMM +++
chain_HMM = alva_MCMC.chain_HMM_node(mmm, 
                                     total_node,
                                     total_path,
                                     node_r,
                                     node_angle_max,
                                     seed_xx,
                                     seed_yy,
                                     seed_angle = None, #seed_path in all directions or in specific seed_angle
                                     chain_level = 1.02,
                                     )
###
real_chain_ii, real_chain_xx, real_chain_yy = chain_HMM[0:3]
seed_node_xx, seed_node_yy = chain_HMM[3:5]
demo_scope = chain_HMM[-1]
###
seed_node_path_xx, seed_node_path_yy, seed_node_path_pp = demo_scope[0:3]
seed_node_path_path0max = demo_scope[3]
seed_node_path_path0_xx, seed_node_path_path0_yy, seed_node_path_path0_pp = demo_scope[4:-2]
seed_node_path_path0_xx_seed, seed_node_path_path0_yy_seed = demo_scope[-2:]
### AlvaHMM ---
###############
### plotting
figure_name = '{:}{:}'.format(section_name, image_name)
file_suffix = '.png'
save_figure = os.path.join(saving_dir_path, file_name + figure_name + file_suffix)
AlvaFontSize = 21
wX = 2
wY = 1
wxx = 9 # window_width
figure = plt.figure(figsize = (wX*wxx, wY*wxx*(raw_pixel_y / raw_pixel_x)))
for wn in iWatcher(range(wX*wY)):
    if (wn == 0):
        window = figure.add_subplot(wY, wX, wn + 1)
        window.set_title('total seeds = {:}'.format(len(seed_xx)), fontsize = AlvaFontSize)
        ### all seeding_point and their tracking
        for i in range(len(seed_xx)):
            window.plot(seed_node_xx[i], seed_node_yy[i],
                        '.', markersize = node_r / 1, 
                        markerfacecolor = 'none', markeredgecolor = 'black', markeredgewidth = 0.25)
            window.plot(seed_node_xx[i, 0], seed_node_yy[i, 0],
                        '*', markersize = 10, markerfacecolor = 'none', markeredgecolor = 'blue')
        ### real_chain
        for i in range(real_chain_ii.shape[0]):
            window.plot(real_chain_xx[i][real_chain_ii[i]], real_chain_yy[i][real_chain_ii[i]],
                        'o', markersize = node_r / 0.8, markerfacecolor = 'none',
                        markeredgecolor = 'red', markeredgewidth = 0.25)
        ###
        window.imshow(mmm, cmap = channel_color + 's', interpolation = 'nearest')
        window.grid(True)
    if (wn == 1):
        ### connect_chain_image +++
        connect_chain_xx = np.array([], dtype = np.int64)
        connect_chain_yy = np.array([], dtype = np.int64)
        for i in range(len(real_chain_ii)):
            point_xx = np.array(real_chain_xx[i][real_chain_ii[i]], dtype = np.int64)
            point_yy = np.array(real_chain_yy[i][real_chain_ii[i]], dtype = np.int64)
            ###
            pixel_line_xx, pixel_line_yy = alva_MCMC.connect_r_point_by_pixel(point_xx, point_yy, node_r)
            connect_chain_xx = np.append(connect_chain_xx, pixel_line_xx)
            connect_chain_yy = np.append(connect_chain_yy, pixel_line_yy)
        ###
        new_mmm = np.zeros([total_pixel_y, total_pixel_x], dtype = np.int64)
        new_mmm[connect_chain_yy, connect_chain_xx] = 1
        ###
        ### saving_new_mmm +++
        file_suffix = '.jpg'
        saving_data_file = file_name + section_name
        mpl.image.imsave(os.path.join(saving_dir_path, saving_data_file + file_suffix), new_mmm)
        ###
        file_suffix = '.npy'
        np.save(os.path.join(saving_dir_path, saving_data_file + file_suffix), new_mmm)
        ### saving_new_mmm ---
        ###         
        window = figure.add_subplot(wY, wX, wn + 1)
        window.set_title('draft_way (selected seed)', fontsize = AlvaFontSize) 
        ###
        window.imshow(new_mmm, cmap = channel_color + 's', interpolation = 'nearest')
        window.set_xlabel('X (pixel)', fontsize = AlvaFontSize)
        window.set_ylabel('Y (pixel)', fontsize = AlvaFontSize)
        window.set_xticklabels(np.int64(window.get_xticks()), fontsize = AlvaFontSize*0.8, rotation = 0) 
        window.set_yticklabels(np.int64(window.get_yticks()), fontsize = AlvaFontSize*0.8)
    ### 
###
figure.tight_layout()
plt.savefig(save_figure, dpi = 128, bbox_inches = 'tight')
plt.show()
##################
```
